```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE


#IMPORT DATA
data = pd.read_csv('irisdata.txt', sep=" ", names=['sl', 'sw', 'pl', 'pw'])
X = data

#IMPORTING KMEANS
def kmeans(X, K):
    # Keep count and track of past cent to see progress of k means
    previous_cent = []
    previous_label = []
    # Next step is to randomly choose the centroids

    index-no = np.random.choice(X.shape[0], K)
    print(index-no)
    cent = X[index-no]
    print(cent)
    previous_cent.append(cent)
    while True:
      
       # Repeating the above steps until there is minimal or no change in the centroid 

        # finding the distance between cent and all the data points

        labels = np.argmin(cdist(X, cent), axis=1) # Centroid with the minimum Distance
        previous_label.append(labels)

        new_cent = np.array([X[labels == i].mean(axis=0) for i in range(K)])
        previous_cent.append(new_cent)

        if np.all(cent == new_cent):
            break
        cent = new_cent

    return cent, labels, previous_cent, previous_label


#Initializing pca into TSNE
pca = TSNE(2) 
# Setting the n-components count for TSNE to 2


# Fit and Transform the data
df = pca.fit_transform(X)

# Defining our function
cent, labels, cent_hist, labels_hist = kmeans(df, 3)

for i in range(len(labels)):
    labels[i] += 1
print(labels)

#Resulting file
result = open('iris-output-part1.txt', 'w')
result.writelines( "%s\n" % i for i in labels )
result.close()

# previous plot

plt.scatter(X['sl'], X['sw'])
plt.title('Iris - sl vs Width')
plt.show()

# recent plot

plt.scatter(X['sl'], X['sw'], c=labels, cmap='tab20b')
plt.title('Iris - sl vs Width - Clustered')
plt.show()
```
